```python
import pandas as pd
import numpy as np
import re
```

```python
data=pd.read_csv('bangalore zomato.csv')
```

```python
data.columns
```

```python
data['cuisines0']= data['cuisines'].str.split(',').str[2]
```

```python
data['cuisines0']= data['cuisines'].apply(lambda x: x.split(',')[0])
```

```python
data['cuisines0']
```

```python
data['cuisines']
```

```python
data['cuisines']= data['cuisines'].astype('str')
```

```python
lists=[]
data.index= range(data.shape[0])
for i in data.index:
    array_split= re.split(',',data['cuisines'][i])
    for items in array_split:
        lists.append(items)
    
```

```python
array_split
```

```python
lists
```

```python
pd.Series(data=lists).value_counts()[:30]
```

```python
likes=[]
for u in data.index:
    re.split(',',data['cuisines'][u])
    for i in re.split(',',data['cuisines'][u]):
        likes.append(i)
        
```

```python
pd.Series(data=likes).value_counts()[:25]
```
